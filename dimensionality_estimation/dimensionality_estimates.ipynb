# Dimensionality estimates

This notebook aggregates the results from which the dimensionality estimates are plotted.



```python
import matplotlib.pyplot as plt
import json
from src import helpers_dimensionality
```

## Load the files

```python
with open('results/table_1.json') as f:
    table_1 = json.load(f)

with open('results/table_2.json') as f:
    table_2 = json.load(f)
    
with open('results/table_3.json') as f:
    table_3 = json.load(f)
    
with open('results/figure_1.json') as f:
    figure_1 = json.load(f)

with open('results/figure_2.json') as f:
    figure_2 = json.load(f)
    
with open('results/table_1_resnet_pretrained.json') as f:
    table_1_resnet = json.load(f)
    
with open('results/figure_1.json') as f:
    figure_1 = json.load(f)

with open('results/figure_2.json') as f:
    figure_2 = json.load(f)
```

```python
resnet_rand_array, resnet_rand_background, _ = helpers_dimensionality.filter_dimensions(table_1['resnet_random'])

# array instances
black_resnet_rand, _, _ = helpers_dimensionality.filter_dimensions(figure_1['black']['resnet_random'])
blue_resnet_rand, _, _ = helpers_dimensionality.filter_dimensions(figure_1['blue']['resnet_random'])
large_resnet_rand, _, _ = helpers_dimensionality.filter_dimensions(figure_1['large']['resnet_random'])
small_resnet_rand, _, _ = helpers_dimensionality.filter_dimensions(figure_1['small']['resnet_random'])

# background instances
_, forest_resnet_rand, _ = helpers_dimensionality.filter_dimensions(figure_2['forest']['resnet_random'])
_, field_resnet_rand, _ = helpers_dimensionality.filter_dimensions(figure_2['field']['resnet_random'])


```

```python
fig, ax = plt.subplots(2,1, figsize = (10, 6))
fig.subplots_adjust(hspace=0.4)


ax[0].boxplot([resnet_rand_array, black_resnet_rand, blue_resnet_rand, large_resnet_rand, small_resnet_rand])
ax[0].set_xticks([1, 2, 3, 4, 5]) # values
ax[0].set_xticklabels(['Overall', 'Black', 'Blue', 'Large', 'Small']) 
ax[0].set_ylabel('Dimensionality')
ax[0].set_title('Estimated dimensions : array')


ax[1].boxplot([resnet_rand_background, forest_resnet_rand, field_resnet_rand])
ax[1].set_title('Estimated dimensions : background')
ax[1].set_xticks([1, 2, 3]) # values
ax[1].set_xticklabels(['Overall', 'Forest', 'Field']) 
ax[1].set_ylabel('Dimensionality')
plt.suptitle('Dimensionality estimates in the training dataset.')

plt.savefig('figs/dimensionality_estimates.pdf')
plt.show()
```
